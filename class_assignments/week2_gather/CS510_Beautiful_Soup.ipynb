```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
%matplotlib inline
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from pylab import rcParams
```

```python
url = "http://www.hubertiming.com/results/2017GPTR10K"
html = urlopen(url)
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
# Get the title
title = soup.title
print(title)
```

```python
# Print out the text
text = soup.get_text()
#print(soup.text)
```

```python
all_links = soup.find_all('a')
print(all_links)
print("\n")
for link in all_links:
    print(link.get("href"))
```

```python
rows = soup.find_all('tr')
print(rows[:10])
```

```python
for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)
```

```python
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, 'lxml').get_text()
print(cleantext)
```

```python
list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean_regex = re.compile('<.*?>')
    clean_cells = re.sub(clean_regex, '', str_cells)
    list_rows.append(clean_cells)
print(clean_cells)
type(clean_cells)
```

```python
df = pd.DataFrame(list_rows)
df.head(10)
```

```python
df1 = df[0].str.split(',', expand = True)
df1.head(10)
```

```python
df1[0] = df1[0].str.strip('[')
df1[0] = df1[0].str.strip(']')
df1[1] = df1[1].str.strip(']')
df1.head(10)
```

```python
all_headers = []
col_headers = soup.find_all('th')
print(col_headers)
str_col_headers = str(col_headers)
clean_col_headers = BeautifulSoup(str_col_headers, 'lxml').get_text()
all_headers.append(clean_col_headers)
print(all_headers)
```

```python
df2 = pd.DataFrame(all_headers)
df2.head()
```

```python
df3 = df2[0].str.split(',', expand = True)
df3.head()
```

```python
frames = [df3, df1]
df4 = pd.concat(frames)
df4.head(10)
```

```python
df5 = df4.rename(columns=df4.iloc[0])
df5.head()
```

```python
df5.info()
df5.shape
```

```python
df6 = df5.dropna(axis=0, how = 'any')
df6.info()
df6.shape
```

```python
df7 = df6.drop(df6.index[0])
df7.head()
```

```python
df7.rename(columns={'[Place':'Place'}, inplace=True)
df7.rename(columns={' Team]':'Team'}, inplace=True)
df7.head()
```

```python
df7['Team'] = df7['Team'].str.strip(']')
df7 = df7.replace({r'\r\n':''}, regex=True)
df7 = df7.replace({r'\n':''}, regex=True)
df7.head()
```

```python
time_list = df7[' Time'].tolist()
time_mins = []
time_component_array = []
for time in time_list:
    time_component_array = time.split(':')
    if len(time_component_array) == 3:
        math = (int(time_component_array[0])*3600 + int(time_component_array[1])*60 + int(time_component_array[2]))/60
        math = round(math, 2)
    else:
        math = (int(time_component_array[0])*60 + int(time_component_array[1]))/60
        math = round(math, 2)
    time_mins.append(math)
```

```python
df7['Runner_mins'] = time_mins
df7.head()
```

```python
df7.describe(include = [np.number])
```

```python
rcParams['figure.figsize'] = 15, 5
df7.boxplot(column='Runner_mins')
plt.grid(True, axis='y')
plt.ylabel('Time')
plt.xticks([1], ['Runners'])
```

```python
x = df7['Runner_mins']
ax = sns.displot(x, kde=True, rug=False, color='m', bins = 25)
plt.show()
```

```python
f_fuko = df7.loc[df7[' Gender']==' F']['Runner_mins']
m_fuko = df7.loc[df7[' Gender']==' M']['Runner_mins']
sns.displot(f_fuko, kde=True, rug=False, label='Female')
sns.displot(m_fuko, kde=True, rug=False, label='Male')
plt.legend()
```

```python
g_stats = df7.groupby(" Gender", as_index = True).describe()
print(g_stats)
```
