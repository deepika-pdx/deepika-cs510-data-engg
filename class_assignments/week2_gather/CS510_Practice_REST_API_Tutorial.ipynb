```python
endpoint = "https://api.nasa.gov/mars-photos/api/v1/rovers/curiosity/photos"
nasa_query_params = {"api_key":"DEMO_KEY", "sol":"1000"}
nasa_response = requests.get(endpoint, nasa_query_params)
nasa_response.headers.get("Content-Type")
nasa_response.json()
```

```python
mars_rover_photos = nasa_response.json()["photos"]
print(f"Found {len(mars_rover_photos)} photos")
```
