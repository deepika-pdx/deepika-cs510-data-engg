```python
import pandas as pd
import numpy as np
import math
```

```python
# Reading the Census data csv file into a pandas dataframe
df = pd.read_csv('G:\\USA\\Masters\\Graduate\\Courses\\CS 510 Data Engg\\Class_Assignments\\Week7\\Data\\acs2017_census_tract_data.csv')
df.head()
```

```python
# Converting the poverty percentages of each county into total number of people under poverty category
def calculateTotalPoverty(poverty, total_pop):
    return ((poverty*total_pop)/100)

df['TotalPoverty'] = df.apply(lambda each_row: calculateTotalPoverty(each_row['Poverty'], each_row['TotalPop']), axis=1)
df.head()
```

```python
# Calculating aggregate income of each county 
def calculateAggregateIncome(income_per_capita, total_pop):
    return (income_per_capita*total_pop)

df['AggregateIncome'] = df.apply(lambda each_row: calculateAggregateIncome(each_row['IncomePerCap'], each_row['TotalPop']), axis=1)
df.head()
```

```python
# Creating a empty County_Info dataframe
County_Info = pd.DataFrame(columns=['ID', 'Name', 'State', 'Population', 'Poverty', 'PerCapitaIncome'])
County_Info.head()
```

```python
# Calculating the Name, State, Total Population, Aggregate Income and Total Poverty of each county
state_list = []
county_list = []
population_list = []
income_list = []
total_poverty_per_county = []

df2 = df.groupby(['State', 'County'])

for state_county, county_group in df2:
    state_county_array = str(state_county).replace("(", "").replace(")", "").replace("'", "").split(",")
    state_list.append(state_county_array[0])
    county_list.append(state_county_array[1])
    population_list.append(county_group['TotalPop'].sum())
    income_list.append(county_group['AggregateIncome'].sum())
    total_poverty_per_county.append(county_group['TotalPoverty'].sum())
    
County_Info['Name'] = county_list
County_Info['State'] = state_list
County_Info['Population'] = population_list
County_Info['AggregateIncomePerCounty'] = income_list
County_Info['TotalPovertyPerCounty'] = total_poverty_per_county
County_Info['ID'] = County_Info['ID'].index + 1

County_Info.head()
```

```python
# Calculating the poverty percentage of each county
def calculatePovertyPercentPerCounty(totalpoverty, population):
    return ((totalpoverty/population)*100)
County_Info['Poverty'] = County_Info.apply(lambda each_row: calculatePovertyPercentPerCounty(each_row['TotalPovertyPerCounty'], each_row['Population']), axis=1)
County_Info.head()
```

```python
# Calculating the per capita income of each county
def calculatePerCapitaIncomePerCounty(aggregate_income, population):
    return (aggregate_income/population)
County_Info['PerCapitaIncome'] = County_Info.apply(lambda each_row: calculatePerCapitaIncomePerCounty(each_row['AggregateIncomePerCounty'], each_row['Population']), axis=1)
County_Info.head()
```

```python
# Dropping the 'AggregateIncomePerCounty' and 'TotalPovertyPerCounty' columns
County_Info = County_Info.drop(columns=['AggregateIncomePerCounty','TotalPovertyPerCounty'])
County_Info.head()
```
